```python
Batters = pd.read_csv("data/Batters.csv")

Batters.head()
```

```python
Batters.describe()
```

```python
# Identify players primary role
positions = ['Appearances_G_p', 'Appearances_G_c', 'Appearances_G_1b',
       'Appearances_G_2b', 'Appearances_G_3b', 'Appearances_G_ss',
       'Appearances_G_lf', 'Appearances_G_cf', 'Appearances_G_rf',
       'Appearances_G_of']

Batters['position'] = Batters[positions].idxmax(axis=1).str.slice(14)
Batters = Batters.drop(Batters[Batters.position == "p"].index)

# Counts do not match R, weird?
print(Batters[['playerID', 'position']].groupby(['position']).agg(['count']))

Batters = pd.concat([Batters, pd.get_dummies(Batters[['position']], prefix=['position'])], axis=1,sort=False)
print(Batters.head())
```
